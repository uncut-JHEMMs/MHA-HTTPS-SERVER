```python
import matplotlib.pyplot as plt
import matplotlib.dates
import csv
import numpy as np
from datetime import datetime
```

```python
req_number = []
total_time  = []
time_connect = []
with open("../latency/latency.csv" , 'r' , newline='') as csvfile:
    reader = csv.reader(csvfile)
    next(reader)
    for row in reader:
        row = row[0].split('\t')
        req_number.append(int(row[0]))
        total_time.append(float(row[7]))
        time_connect.append(float(row[2]))
```

```python
plt.style.use('_mpl-gallery')
colors = np.random.uniform(20 , 80 , len(req_number))
fig , ax = plt.subplots()
ax.scatter(req_number , total_time , s=10 , c=colors)
fig.set_dpi(200)
plt.show()
```

```python
plt.style.use('_mpl-gallery')
colors = np.random.uniform(20 , 80 , len(req_number))
fig , ax = plt.subplots()
ax.scatter(req_number , time_connect , s=40 , c=colors)
fig.set_dpi(200)
plt.show()
```

```python
time = []
cpu_usage  = []
mem_usage = []
with open("../usages/cpu_usage_log.csv" , 'r' , newline='') as csvfile:
    reader = csv.reader(csvfile)
    next(reader)
    for row in reader:
        row = row[0].split('\t')
        st = row[0].split(' ')
        
        time.append(datetime( int(st[3]) , 4 , int(st[1]) , int(st[4][0:2]) , int(st[4][3:5]) , int(st[4][6:8]) ))
        cpu_usage.append(float(row[1]))
        mem_usage.append(float(row[2]))
```

```python
#plt.style.use('_mpl-gallery')
colors = np.random.uniform(20 , 80 , len(req_number))
plt.figure(figsize=(10 , 10))
plt.plot(time , cpu_usage)
plt.plot(time , mem_usage)
plt.gcf().autofmt_xdate()
plt.show()
```
